```python
""" Create a Python class called `BankAccount` with private attributes for the account balance (`__balance`)
and account number (`__account_number`). Provide methods for depositing and withdrawing money
"""

class BankAccount:
    def __init__(self, balance, account_number):
        self.__balance = balance
        self.__account_number = account_number
        
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawal of ${amount} successful.")
        else:
            print("Withdrawal failed. Insufficient funds.")
           
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposit of ${amount} successful.")
        else:
            print("Deposit failed. Please deposit a positive amount.")
            
    def get_balance(self):
        return self.__balance
    
    def get_account_number(self):
        return self.__account_number


account = BankAccount(1000, "1234567890")
print("Initial Balance:", account.get_balance())  
account.withdraw(200)  
print("Updated Balance:", account.get_balance())  
account.deposit(500)   
print("Updated Balance:", account.get_balance())              
```

```python
"""Discuss the use of abstract methods and classes in achieving polymorphism in Python. Provide an example
using the `abc` module
"""

"""
Common platform: Abstract classes provide a way to define a common interface that all subclasses must adhere to. 
This allows you to enforce a certain structure or behavior across multiple classes.


Code resueability:Abstract class ensure that the abstract function is defined by child class whuch eventually leads to code reuseability.


"""

#to use abstract methods in python we have to use ABC module 
 
from abc import ABC, abstractmethod

class task(ABC):  
    @abstractmethod
    def process(self):
        pass

class work(task):
    def process(self, x, y):  
        print("Sum of numbers is:", x + y)

obj_work = work()  
obj_work.process(10, 20)
```

```python
"""Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and
providing methods to deposit and withdraw funds.
"""
from abc import ABC, abstractmethod

class Bank(ABC):  
    
    def __init__(self, account_number, initial_balance=0):
        self._account_number = account_number
        self._balance = initial_balance  

    @abstractmethod
    def deposit(self, amount):
        pass
    
    @abstractmethod
    def withdraw(self, amount):
        pass

    def balance_check(self):
        return self._balance

class BankWorking (Bank):  
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print("Amount to be deposited:",amount,"New balance is:",self._balance)
        else:
            print("Deposit amount must be greater than zero.")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print("Withdrew amount is:",amount, "New balance is :", self._balance)
        else:
            print("Withdrawal amount must be greater than zero and less than or equal to the balance.")


bank_account = BankWorking(account_number="123456789", initial_balance=1000)
print("Initial balance:", bank_account.balance_check())

bank_account.deposit(500)
bank_account.withdraw(200)
print("Current balance:", bank_account.balance_check())
        
```

```python
"""Create a Python class for a vehicle system and demonstrate abstraction by defining common methods
(e.g., `start()`, `stop()`) in an abstract base class.
"""

from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def model_details(self):
        pass 
    
    @abstractmethod
    def start(self):
        pass
    
    @abstractmethod 
    def stop(self):
        pass 
  
    def show(self):
        pass

class VehicleWorking(Vehicle):
    def model_details(self, model_name, reg_id, color, owner):
        self.model_name = model_name
        self.reg_id = reg_id
        self.color = color
        self.owner = owner
        
    def start(self, startingprice):
        self.startingprice = startingprice
        print("Vehicle starting price:", self.startingprice)

    def stop(self, stopingprice):
        self.stopingprice = stopingprice
        print("Vehicle stopping price:", self.stopingprice)
        
    def show(self):
        print("Model Number:", self.model_name)
        print("Registration ID:", self.reg_id)
        print("Color:", self.color)
        print("Owner:", self.owner)
        print("Starting Price:", self.startingprice)
        print("Stopping Price:", self.stopingprice)
        

car = VehicleWorking()
car.model_details("Toyota Camry", "12345", "Black", "John Doe")
car.start(451000)
car.stop(689900)
car.show()
```

```python
"""Create a Python class for a computer system, demonstrating abstraction by defining common methods
(e.g., `power_on()`, `shutdown()`) in an abstract base class
"""

from abc import ABC, abstractmethod

class Base(ABC):
    def __init__(self, name, pcid):
        self.name = name
        self.pcid = pcid
    
    @abstractmethod
    def pc_details(self):
        pass
    
    @abstractmethod
    def power_on(self):
        pass
    
    @abstractmethod
    def shutdown(self):
        pass

class Child(Base):
    def __init__(self, name, pcid):
        super().__init__(name, pcid)
    
    def pc_details(self):
        print("Name:", self.name)
        print("PC ID:", self.pcid)
    
    def power_on(self, starting_time):
        print(self.name, "is powering on at", starting_time)
    
    def shutdown(self, shutdown_time):
        print(self.name, "is shutting down at", shutdown_time)


computer = Child("My Computer", "PC123")
computer.pc_details()
computer.power_on("9:00 AM")
computer.shutdown("6:00 PM")
```

```python
"""Create a Python class for a library system, implementing abstraction by defining common methods (e.g.,
`add_book()`, `borrow_book()`) in an abstract base class
"""

from abc import ABC, abstractmethod

class Library(ABC):
    def __init__(self, bookname, bookid):
        self.bookname = bookname
        self.bookid = bookid
        
    @abstractmethod
    def show(self):
        pass
    
    @abstractmethod    
    def add_book(self):
        pass
    
    @abstractmethod
    def borrow_book(self):
        pass 

class MaintainLibrary(Library):
    def show(self):
        print("Book Name:", self.bookname)
        print("Book ID:", self.bookid)
    
    def add_book(self, new_bookname, new_bookid):
        print("Adding Book:", new_bookname)
        print("Book ID:", new_bookid)
    
    def borrow_book(self, borrower_name):
        print("Borrowing Book:", self.bookname)
        print("Borrower:", borrower_name)


library = MaintainLibrary("Python Programming", "PY101")
library.show()
library.add_book("Data Structures and Algorithms", "DSA202")
library.borrow_book("Bt")
     
```

```python
"""Create a Python class hierarchy for a music player system, using composition to represent playlists and
songs
"""


class Song:
    def __init__(self, song):
        self.song = song

    def take_song(self):
        return self.song

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

class MusicPlayer:
    def __init__(self, songs, playlists):
        self.songs = songs
        self.playlists = playlists

    def add_song(self, song):
        self.songs.append(song)

    def add_playlist(self, playlist):
        self.playlists.append(playlist)


song1 = Song("Song 1")
song2 = Song("Song 2")

playlist1 = Playlist("My Playlist")
playlist1.add_song(song1)
playlist1.add_song(song2)

player = MusicPlayer([song1, song2], [playlist1])
```

```python
""" Create a Python class for a car, using composition to represent components like the engine, wheels, and
transmission
"""


class Engine:
    def __init__(self, name):
        self.name = name

class Wheels:
    def __init__(self, count):
        self.count = count

class Transmission:
    def __init__(self, type):
        self.type = type

class Car:
    def __init__(self, carname, engine, wheels, transmission):
        self.carname = carname
        self.engine = engine
        self.wheels = wheels
        self.transmission = transmission

class CarDetails:
    def __init__(self, car):
        self.car = car

    def show(self):
        print("Car name:", self.car.carname)
        print("Engine:", self.car.engine.name)
        print("Number of wheels:", self.car.wheels.count)
        print("Transmission type:", self.car.transmission.type)


engine = Engine("VT400")
wheels = Wheels(6)
transmission = Transmission("Auto")


car = Car("Ferrari", engine, wheels, transmission)


car_details = CarDetails(car)


car_details.show()
```

```python
"""Create a Python class for a university course, using composition to represent students, instructors, and
course materials
"""


class Student:
    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno

class Instructor:
    def __init__(self, name, instructor_id):
        self.name = name
        self.instructor_id = instructor_id

class Materials:
    def __init__(self, materials):
        self.materials = materials

class UniversityCourse:
    def __init__(self, student, instructor, materials):
        self.student = student
        self.instructor = instructor
        self.materials = materials
    
    def show(self):
        print("Student name:", self.student.name)
        print("Student roll number:", self.student.rollno)
        print("Instructor name:", self.instructor.name)
        print("Instructor ID:", self.instructor.instructor_id)
        print("Materials provided:", self.materials.materials)


student = Student("John Doe", 123)
instructor = Instructor("Dr. Smith", 456)
study_materials = Materials(["Science Book", "Computer Book", "Math Book"])

course = UniversityCourse (student, instructor, study_materials)
course.show()
```

```python
"""Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes,
and ingredients
"""

class Dish:
    def __init__(self, dish_menu):
        self.dish_menu = dish_menu

class Menu:
    def __init__(self, menu_items):
        self.menu_items = menu_items

class Ingredients:
    def __init__(self, items):
        self.items = items

class Restaurant:
    def __init__(self, dish, menu, ingredients):
        self.dish = dish
        self.menu = menu
        self.ingredients = ingredients
    
    def show(self):
        print("Dish menu:", self.dish.dish_menu)
        print("Menu items:", self.menu.menu_items)
        print("Ingredients:", self.ingredients.items)


dish_list = ['Egg Rice', 'Paneer Rice']        
obj_dish = Dish(dish_list)

food_list = ['Pineapple Cake', 'Potato Chips', 'Biryani', 'Chicken']
obj_menu = Menu(food_list)

ingredient_list = ['Eggs', 'Rice', 'Paneer', 'Pineapple', 'Potatoes', 'Chicken']
obj_ingredients = Ingredients(ingredient_list)

restaurant = Restaurant(obj_dish, obj_menu, obj_ingredients)
restaurant.show()
```

```python
"""How can you achieve flexibility in composed objects by allowing them to be replaced or modified
dynamically at runtime?
"""

from abc import ABC, abstractmethod

class Base(ABC):
    @abstractmethod
    def show(self):
        pass 

class Child(Base):
    def __init__(self, x):
        self.x = x

    def show(self):
        print("Value of x:", self.x)


obj = Child(10)
obj.show()  


obj.x = 20
obj.show() 
```

```python
"""Create a Python class for a social media application, using composition to represent users, posts, and
comments.
"""

class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password
        self.posts = []

    def create_post(self, content):
        post = Post(content)
        self.posts.append(post)

    def display_posts(self):
        if self.posts:
            print("Posts by", self.username)
            for post in self.posts:
                print(post.content)
        else:
            print("No posts found.")

class Post:
    def __init__(self, content):
        self.content = content
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_comments(self):
        if self.comments:
            print("Comments on the post:")
            for comment in self.comments:
                print(comment)
        else:
            print("No comments yet.")

class SocialMediaApp:
    def __init__(self):
        self.users = {}

    def signup(self, username, password):
        if username not in self.users:
            self.users[username] = User(username, password)
            print("Signup successful!")
        else:
            print("Username already exists. Please choose a different username.")

    def login(self, username, password):
        if username in self.users:
            if self.users[username].password == password:
                print("Login successful!")
                return self.users[username]
            else:
                print("Incorrect password. Please try again.")
        else:
            print("User not found. Please signup.")


app = SocialMediaApp()

while True:
    print("\nMenu:")
    print("1. Signup")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        username = input("Enter username: ")
        password = input("Enter password: ")
        app.signup(username, password)
    elif choice == '2':
        username = input("Enter username: ")
        password = input("Enter password: ")
        user = app.login(username, password)
        if user:
            while True:
                print("\nUser Menu:")
                print("1. Create Post")
                print("2. Display Posts")
                print("3. Logout")
                user_choice = input("Enter your choice: ")
                if user_choice == '1':
                    content = input("Enter post content: ")
                    user.create_post(content)
                elif user_choice == '2':
                    user.display_posts()
                elif user_choice == '3':
                    break
                else:
                    print("Invalid choice.")
    elif choice == '3':
        print("Exiting")
        break
    else:
        print("Invalid choice. Please choose again.")
       
```
