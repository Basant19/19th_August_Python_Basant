```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Smokes'], bins=10, kde=True)
plt.title('Smokes Distribution')
plt.xlabel('Number of Cigarettes Smoked Per Day')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Alkhol'], bins=10, kde=True)
plt.title('Alcohol Consumption Distribution')
plt.xlabel('Alcohol Consumption Level')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df[['Age', 'Smokes', 'AreaQ', 'Alkhol', 'Result']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Result', y='Age', data=df)
plt.title('Age vs. Lung Cancer')
plt.xlabel('Lung Cancer (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
#it contains outliers 
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Result', y='Smokes', data=df, estimator='mean')
plt.title('Bar Plot: Average Smokes vs. Lung Cancer')
plt.xlabel('Lung Cancer (0: No, 1: Yes)')
plt.ylabel('Average Number of Cigarettes Smoked Per Day')
plt.show()
#person who smokes have high chance of cancer  
```
