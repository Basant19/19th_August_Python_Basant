```python
''' 1.6. Write a program to display the appropriate message as per the color of signal(RED-Stop/Yellow-Stay/ Green-Go) at the road crossing.  
'''
def tsm(color):
    color = color.lower()
    if color == "red":
        return "RED = Stop"
    elif color == "yellow":
        return "YELLOW = Stay"
    elif color == "green":
        return "GREEN = Go"
    else:
        return "Invalid color. Please enter RED, YELLOW, or GREEN."
def main():
    color = input("Enter the traffic signal color (RED, YELLOW, GREEN): ")
    message = tsm(color)
    print(message)
main()    
```

```python
'''1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*)
'''
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

def main():
    print("Simple Calculator")
    print("Select operation:")
    print("1. Add (+)")
    print("2. Subtract (-)")
    print("3. Multiply (*)")
    print("4. Divide (/)")

    while True:
      
        choice = input("Enter choice (+, -, *, /): ")

        if choice in ('+', '-', '*', '/'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter numeric values.")
                continue

            if choice == '+':
                print(f"{num1} + {num2} = {add(num1, num2)}")

            elif choice == '-':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")

            elif choice == '*':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")

            elif choice == '/':
                print(f"{num1} / {num2} = {divide(num1, num2)}")
        else:
            print("Invalid input. Please enter a valid operator.")

        
        next_calculation = input("Do you want to perform another calculation? (yes/no): ")
        if next_calculation.lower() != 'yes':
            break
main()
```

```python
'''1.9.  Write a program to find the factors of a whole number  using a while loop. 
'''
def facto(num):
    factors = []
    i = 1
    while i <= num:
        if num % i == 0:
            factors.append(i)
        i += 1
    
    return factors


def main ():
    num = int(input("Enter a whole number: "))
    if num < 1:
        print("Please enter a positive whole number.")
        return
    factors = facto(num)
    print(f"The factors of {num} are: {factors}")


main()
```

```python
'''1.10.  Write a program to find the sum of all the positive numbers entered by the user. As soon as the user  enters a negative number, stop taking in any further input from the user and display the sum .
'''
def main():
    total_sum = 0  

    while True:
        try:
            
            num = float(input("Enter a positive number (negative number to stop): "))
            
            if num < 0:
                break  
            total_sum += num  

        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    
    print(f"The sum of all positive numbers entered is: {total_sum}")


main()
```

```python
''' 1.11.   Write a program to find prime numbers between 2 to 100 using nested for loops. 
'''
def main():
    print("Prime numbers between 2 and 100 are:")

    for num in range(2, 101):
      
        is_prime = True

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num, end=" ")


main()
```

```python
'''Write the  programs for the followingM 
. AccePt the marks of the student in five major subjects and disPlay the sameS 
. Calculate the sum of the marks of all subjects.Divide the total marks by number of subjects (i.e. 5), calculate  Tercentage = total marks/5 and disPlay the PercentageS 
. Find the grade of the student as Ter the following criteria . Hint: Use Match & case for this.:
Percentage >85 A
Per <85 and per >=75 B
Per <75 and per >50  C
Per 30 and per <=50  D
Per <30      reappear
'''
def main():
    
    marks = []
    subjects = ["Math", "Science", "English", "History", "Geography"]
    
    for subject in subjects:
        while True:
            try:
                mark = float(input(f"Enter marks for {subject}: "))
                if mark < 0 or mark > 100:
                    print("Please enter a valid mark between 0 and 100.")
                else:
                    marks.append(mark)
                    break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    
    
    print("\nMarks of the student:")
    for subject, mark in zip(subjects, marks):
        print(f"{subject}: {mark}")
    
    
    total_marks = sum(marks)
    percentage = total_marks / 5
    
   
    print(f"\nTotal Marks: {total_marks}")
    print(f"Percentage: {percentage:.2f}%")
    
   
    def determine_grade(percentage):
        match percentage:
            case p if p > 85:
                return "A"
            case p if p >= 75:
                return "B"
            case p if p > 50:
                return "C"
            case p if p <= 50 and p >= 30:
                return "D"
            case _:
                return "Reappear"
    
    grade = determine_grade(percentage)
    print(f"Grade: {grade}")


main()
```

```python
'''1.13. Write a program for VIBGYOR Spectrum based on their Wavelength using. Wavelength Range:â€¨ 
Violet 400 - 440
Indigo 440 - 460
Blue 460- 500
Green 500- 570 
Yellow 570- 590
Orange 590- 620
Red 620-720
'''


def get_vibgyor_color(wavelength):
    match wavelength:
        case w if 400 <= w < 440:
            return "Violet"
        case w if 440 <= w < 460:
            return "Indigo"
        case w if 460 <= w < 500:
            return "Blue"
        case w if 500 <= w < 570:
            return "Green"
        case w if 570 <= w < 590:
            return "Yellow"
        case w if 590 <= w < 620:
            return "Orange"
        case w if 620 <= w <= 720:
            return "Red"
        case _:
            return "Wavelength out of VIBGYOR range"

def main():
    try:
        wavelength = float(input("Enter the wavelength (in nm): "))
        color = get_vibgyor_color(wavelength)
        print(f"The color for the wavelength {wavelength} nm is: {color}")
    except ValueError:
        print("Invalid input. Please enter a numeric value for the wavelength.")

    main()
```

```python
'''Finding the force of earth and moon 
'''
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2


mass_earth = 5.972e24  
mass_moon = 7.34767309e22  
mass_sun = 1.989e30  
distance_earth_sun = 1.496e11  
distance_moon_earth = 3.844e8 


force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)


force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)


print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")


if force_earth_sun > force_moon_earth:
    stronger_force = "Earth-Sun"
else:
    stronger_force = "Moon-Earth"

print(f"The stronger gravitational force is between: {stronger_force}")


if force_moon_earth > force_earth_sun:
    more_attracted = "Moon is more attracted to Earth"
else:
    more_attracted = "Earth is more attracted to Sun"

print(more_attracted)

if force_moon_earth > force_earth_sun:
    attraction = "The Moon is more attracted to the Earth."
else:
    attraction = "The Earth is more attracted to the Sun."

print(attraction)
```

```python
'''student details 
'''

class Student:
    def __init__(self, name, age, roll_number):
        # Encapsulated attributes
        self.__name = name
        self.__age = age
        self.__roll_number = roll_number


    def get_name(self):
        return self.__name

    
    def set_name(self, name):
        self.__name = name

    
    def get_age(self):
        return self.__age

    
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age must be a positive number.")

    
    def get_roll_number(self):
        return self.__roll_number

    
    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number

    
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age: {self.__age}")
        print(f"Roll Number: {self.__roll_number}")

    
    def update_details(self, name=None, age=None, roll_number=None):
        if name is not None:
            self.set_name(name)
        if age is not None:
            self.set_age(age)
        if roll_number is not None:
            self.set_roll_number(roll_number)
        print("Details updated successfully.")


student1 = Student("bt", 25, "CS10")


print("Initial details of student1:")
student1.display_info()
print()

print("Updating details of student1...")
student1.update_details(name="Jane Doe", age=21, roll_number="CS102")
print()


print("Updated details of student1:")
student1.display_info()
print()


student2 = Student("Ali", 22, "CS13")


print("Initial details of student2:")
student2.display_info()
print()


print("Updating age of student2...")
student2.update_details(age=23)
print()


print("Updated details of student2:")
student2.display_info()
```

```python
'''Library book system
'''

class LibraryBook:
    def __init__(self, name, author):
        self.__name = name
        self.__author = author
        self.__available = True  

    
    def get_name(self):
        return self.__name

  
    def set_name(self, name):
        self.__name = name

    
    def get_author(self):
        return self.__author

    def set_author(self, author):
        self.__author = author

    
    def is_available(self):
        return self.__available

   
    def borrow_book(self):
        if self.__available:
            self.__available = False
            print(f"You have successfully borrowed '{self.__name}' by {self.__author}.")
        else:
            print(f"Sorry, '{self.__name}' is currently unavailable.")

    
    def return_book(self):
        if not self.__available:
            self.__available = True
            print(f"'{self.__name}' by {self.__author} has been successfully returned.")
        else:
            print(f"'{self.__name}' was not borrowed.")




book1 = LibraryBook("The Great Gatsby", "F. Scott Fitzgerald")
book2 = LibraryBook("1984", "George Orwell")


print(f"Is '{book1.get_name()}' available? {'Yes' if book1.is_available() else 'No'}")
print(f"Is '{book2.get_name()}' available? {'Yes' if book2.is_available() else 'No'}")
print()


book1.borrow_book()
book2.borrow_book()
print()


book1.borrow_book()  
print()


book1.return_book()
book2.return_book()
print()


book1.return_book()  
```

```python
'''Baking System
'''
class BankAccount:
    def __init__(self, account_number, account_holder, initial_balance=0):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = initial_balance

    # Getter for account number
    def get_account_number(self):
        return self.__account_number

    # Getter for account holder
    def get_account_holder(self):
        return self.__account_holder

    # Getter for balance
    def get_balance(self):
        return self.__balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}.")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.__balance}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to display account information
    def display_info(self):
        print(f"Account Number: {self.__account_number}")
        print(f"Account Holder: {self.__account_holder}")
        print(f"Balance: ${self.__balance}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, initial_balance=0, interest_rate=0.02):
        super().__init__(account_number, account_holder, initial_balance)
        self.__interest_rate = interest_rate

    # Method to add interest
    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        print(f"Added interest: ${interest}. New balance: ${self.get_balance()}.")


class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, initial_balance=0, overdraft_limit=500):
        super().__init__(account_number, account_holder, initial_balance)
        self.__overdraft_limit = overdraft_limit

    # Method to withdraw money with overdraft
    def withdraw(self, amount):
        if amount > 0:
            if (self.get_balance() + self.__overdraft_limit) >= amount:
                super().withdraw(amount)
            else:
                print("Insufficient funds and overdraft limit exceeded.")
        else:
            print("Withdrawal amount must be positive.")


# Testing the banking system

# Creating instances of SavingsAccount and CheckingAccount
savings_account = SavingsAccount("123456789", "John Doe", 1000, 0.03)
checking_account = CheckingAccount("987654321", "Jane Doe", 500, 200)

# Display initial account information
print("Savings Account Info:")
savings_account.display_info()
print()

print("Checking Account Info:")
checking_account.display_info()
print()

# Perform transactions on SavingsAccount
savings_account.deposit(500)
savings_account.withdraw(200)
savings_account.add_interest()
print()

# Perform transactions on CheckingAccount
checking_account.deposit(300)
checking_account.withdraw(600)  # Should be successful due to overdraft limit
checking_account.withdraw(500)  # Should fail due to insufficient funds
```

```python
'''Restrauant  System 
'''

class MenuItem:
    def __init__(self, name, description, price, category):
        self.__id = id(self)  
        self.__name = name
        self.__description = description
        self.__price = price
        self.__category = category

    
    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

   
    def get_description(self):
        return self.__description

    def set_description(self, description):
        self.__description = description

   
    def get_price(self):
        return self.__price

    def set_price(self, price):
        if price >= 0:
            self.__price = price
        else:
            print("Price cannot be negative.")

   
    def get_category(self):
        return self.__category

    def set_category(self, category):
        self.__category = category

    
    def display_info(self):
        print(f"ID: {self.__id}")
        print(f"Name: {self.__name}")
        print(f"Description: {self.__description}")
        print(f"Price: ${self.__price:.2f}")
        print(f"Category: {self.__category}")

class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, is_spicy):
        super().__init__(name, description, price, category)
        self.__is_spicy = is_spicy

    
    def get_is_spicy(self):
        return self.__is_spicy

    def set_is_spicy(self, is_spicy):
        self.__is_spicy = is_spicy

    
    def display_info(self):
        super().display_info()
        print(f"Spicy: {'Yes' if self.__is_spicy else 'No'}")

class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, size):
        super().__init__(name, description, price, category)
        self.__size = size

    
    def get_size(self):
        return self.__size

    def set_size(self, size):
        self.__size = size

    
    def display_info(self):
        super().display_info()
        print(f"Size: {self.__size}")


pizza = FoodItem("Pepperoni Pizza", "A classic pizza with pepperoni.", 12.99, "Food", True)
coffee = BeverageItem("Coffee", "Hot brewed coffee.", 2.99, "Beverage", "Medium")


print("Menu Item Details:")
pizza.display_info()
print()
coffee.display_info()
print()


pizza.set_price(13.99)
pizza.set_description("A classic pizza with extra pepperoni.")
print("Updated Pizza Details:")
pizza.display_info()
```

```python
'''Hotel Management
'''
class Room:
    def __init__(self, room_number, room_type, rate, availability=True):
        self.__room_id = id(self)  
        self.__room_number = room_number
        self.__room_type = room_type
        self.__rate = rate
        self.__availability = availability

    
    def get_room_number(self):
        return self.__room_number

    def set_room_number(self, room_number):
        self.__room_number = room_number

    
    def get_room_type(self):
        return self.__room_type

    def set_room_type(self, room_type):
        self.__room_type = room_type

    
    def get_rate(self):
        return self.__rate

    def set_rate(self, rate):
        if rate >= 0:
            self.__rate = rate
        else:
            print("Rate cannot be negative.")

   
    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.__room_number} has been booked.")
        else:
            print(f"Room {self.__room_number} is not available.")

   
    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.__room_number}.")
        else:
            print(f"Room {self.__room_number} is not booked.")

    
    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out from room {self.__room_number}.")
        else:
            print(f"Room {self.__room_number} is already available.")

   
    def display_info(self):
        print(f"Room ID: {self.__room_id}")
        print(f"Room Number: {self.__room_number}")
        print(f"Room Type: {self.__room_type}")
        print(f"Rate: ${self.__rate:.2f}")
        print(f"Availability: {'Available' if self.__availability else 'Not Available'}")

class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_living_room):
        super().__init__(room_number, "Suite", rate)
        self.__has_living_room = has_living_room

    def get_has_living_room(self):
        return self.__has_living_room

    def set_has_living_room(self, has_living_room):
        self.__has_living_room = has_living_room

    
    def display_info(self):
        super().display_info()
        print(f"Has Living Room: {'Yes' if self.__has_living_room else 'No'}")

class StandardRoom(Room):
    def __init__(self, room_number, rate, bed_count):
        super().__init__(room_number, "Standard", rate)
        self.__bed_count = bed_count

    
    def get_bed_count(self):
        return self.__bed_count

    def set_bed_count(self, bed_count):
        self.__bed_count = bed_count

    
    def display_info(self):
        super().display_info()
        print(f"Bed Count: {self.__bed_count}")


suite_room = SuiteRoom("101", 200, True)
standard_room = StandardRoom("202", 100, 2)


print("Suite Room Details:")
suite_room.display_info()
print()
print("Standard Room Details:")
standard_room.display_info()
print()


suite_room.book_room()
suite_room.check_in()
suite_room.check_out()
suite_room.display_info()
print()

standard_room.book_room()
standard_room.check_in()
standard_room.check_out()
standard_room.display_info()
```
