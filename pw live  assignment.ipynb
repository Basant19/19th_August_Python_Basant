```python
'''1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*)
'''
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

def menu():
    print("Simple Calculator")
    print("Select operation:")
    print("1. Add (+)")
    print("2. Subtract (-)")
    print("3. Multiply (*)")
    print("4. Divide (/)")

    while True:
        choice = input("Enter choice (+, -, *, /): ")

        if choice in ('+', '-', '*', '/'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter numeric values.")
                continue

            if choice == '+':
                print(f"{num1} + {num2} = {add(num1, num2)}")

            elif choice == '-':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")

            elif choice == '*':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")

            elif choice == '/':
                result = divide(num1, num2)
                print(f"{num1} / {num2} = {result}")
        else:
            print("Invalid input. Please enter a valid operator.")

        next_calculation = input("Do you want to perform another calculation? (yes/no): ")
        if next_calculation.lower() != 'yes':
            break

menu()
```

```python
'''1.9.  Write a program to find the factors of a whole number  using a while loop. 
'''
def facto(num):
    factors = []
    i = 1
    while i <= num:
        if num % i == 0:
            factors.append(i)
        i += 1
    
    return factors


def main ():
    num = int(input("Enter a whole number: "))
    if num < 1:
        print("Please enter a positive whole number.")
        return
    factors = facto(num)
    print(f"The factors of {num} are: {factors}")


main()
```

```python
''' 1.11.   Write a program to find prime numbers between 2 to 100 using nested for loops. 
'''
def main():
    print("Prime numbers between 2 and 100 are:")

    for num in range(2, 101):
      
        is_prime = True

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num, end=" ")


main()
```

```python
'''Finding the force of earth and moon 
'''
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2


mass_earth = 5.972e24  
mass_moon = 7.34767309e22  
mass_sun = 1.989e30  
distance_earth_sun = 1.496e11  
distance_moon_earth = 3.844e8 


force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)


force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)


print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")


if force_earth_sun > force_moon_earth:
    stronger_force = "Earth-Sun"
else:
    stronger_force = "Moon-Earth"

print(f"The stronger gravitational force is between: {stronger_force}")


if force_moon_earth > force_earth_sun:
    more_attracted = "Moon is more attracted to Earth"
else:
    more_attracted = "Earth is more attracted to Sun"

print(more_attracted)

if force_moon_earth > force_earth_sun:
    attraction = "The Moon is more attracted to the Earth."
else:
    attraction = "The Earth is more attracted to the Sun."

print(attraction)
```

```python
'''Baking System
'''
class BankAccount:
    def __init__(self, account_number, account_holder, initial_balance=0):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = initial_balance

   
    def get_account_number(self):
        return self.__account_number

   
    def get_account_holder(self):
        return self.__account_holder

    
    def get_balance(self):
        return self.__balance

   
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}.")
        else:
            print("Deposit amount must be positive.")

 
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.__balance}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

   
    def display_info(self):
        print(f"Account Number: {self.__account_number}")
        print(f"Account Holder: {self.__account_holder}")
        print(f"Balance: ${self.__balance}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, initial_balance=0, interest_rate=0.02):
        super().__init__(account_number, account_holder, initial_balance)
        self.__interest_rate = interest_rate

    
    def add_interest(self):
        interest = self.get_balance() * self.__interest_rate
        self.deposit(interest)
        print(f"Added interest: ${interest}. New balance: ${self.get_balance()}.")


class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, initial_balance=0, overdraft_limit=500):
        super().__init__(account_number, account_holder, initial_balance)
        self.__overdraft_limit = overdraft_limit

    
    def withdraw(self, amount):
        if amount > 0:
            if (self.get_balance() + self.__overdraft_limit) >= amount:
                super().withdraw(amount)
            else:
                print("Insufficient funds and overdraft limit exceeded.")
        else:
            print("Withdrawal amount must be positive.")





savings_account = SavingsAccount("123456789", "John Doe", 1000, 0.03)
checking_account = CheckingAccount("987654321", "Jane Doe", 500, 200)


print("Savings Account Info:")
savings_account.display_info()
print()

print("Checking Account Info:")
checking_account.display_info()
print()


savings_account.deposit(500)
savings_account.withdraw(200)
savings_account.add_interest()
print()


checking_account.deposit(300)
checking_account.withdraw(600)  
checking_account.withdraw(500)  
```

```python
'''Hotel Management
'''

class Room:
    def __init__(self, room_number, room_type, rate, availability=True):
        self.__room_id = id(self)  
        self.__room_number = room_number
        self.__room_type = room_type
        self.__rate = rate
        self.__availability = availability

    
    def get_room_number(self):
        return self.__room_number

    def set_room_number(self, room_number):
        self.__room_number = room_number

    
    def get_room_type(self):
        return self.__room_type

    def set_room_type(self, room_type):
        self.__room_type = room_type

    
    def get_rate(self):
        return self.__rate

    def set_rate(self, rate):
        if rate >= 0:
            self.__rate = rate
        else:
            print("Rate cannot be negative.")

   
    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.__room_number} has been booked.")
        else:
            print(f"Room {self.__room_number} is not available.")

   
    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.__room_number}.")
        else:
            print(f"Room {self.__room_number} is not booked.")

    
    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out from room {self.__room_number}.")
        else:
            print(f"Room {self.__room_number} is already available.")

   
    def display_info(self):
        print(f"Room ID: {self.__room_id}")
        print(f"Room Number: {self.__room_number}")
        print(f"Room Type: {self.__room_type}")
        print(f"Rate: ${self.__rate:.2f}")
        print(f"Availability: {'Available' if self.__availability else 'Not Available'}")

class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_living_room):
        super().__init__(room_number, "Suite", rate)
        self.__has_living_room = has_living_room

    def get_has_living_room(self):
        return self.__has_living_room

    def set_has_living_room(self, has_living_room):
        self.__has_living_room = has_living_room

    
    def display_info(self):
        super().display_info()
        print(f"Has Living Room: {'Yes' if self.__has_living_room else 'No'}")

class StandardRoom(Room):
    def __init__(self, room_number, rate, bed_count):
        super().__init__(room_number, "Standard", rate)
        self.__bed_count = bed_count

    
    def get_bed_count(self):
        return self.__bed_count

    def set_bed_count(self, bed_count):
        self.__bed_count = bed_count

    
    def display_info(self):
        super().display_info()
        print(f"Bed Count: {self.__bed_count}")


suite_room = SuiteRoom("101", 200, True)
standard_room = StandardRoom("202", 100, 2)


print("Suite Room Details:")
suite_room.display_info()
print()
print("Standard Room Details:")
standard_room.display_info()
print()


suite_room.book_room()
suite_room.check_in()
suite_room.check_out()
suite_room.display_info()
print()

standard_room.book_room()
standard_room.check_in()
standard_room.check_out()
standard_room.display_info()
```

```python
'''Fitness club  
'''

import random

class FitnessClub:
    def __init__(self, name, age, membership_type):
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__id = self.generate_id()
        self.__membership_status = True  

    def generate_id(self):
        return random.randint(10000, 99999)

    def get_id(self):
        return self.__id

    def get_membership_status(self):
        return self.__membership_status

    def register_member(self):
        self.__membership_status = True
        print(f"Member Name: {self.name}")
        print(f"ID Number: {self.__id}")
        print("Registered successfully!")

    def renew_membership(self):
        if not self.__membership_status:
            print(f"Membership of {self.name} has been cancelled. Please register again.")
        else:
            print(f"Membership of {self.name} has been renewed.")

    def cancel_membership(self):
        self.__membership_status = False
        print(f"Membership of {self.name} has been cancelled.")


class FamilyMember(FitnessClub):
    def __init__(self, name, age, membership_type, family_size):
        super().__init__(name, age, membership_type)
        self.family_size = family_size

    def add_family_member(self):
        if self.get_membership_status():
            print(f"Added a new family member to {self.name}'s membership. Family size: {self.family_size}")
        else:
            print("Cannot add family members. Membership is not active.")


class IndividualMember(FitnessClub):
    def __init__(self, name, age, membership_type, personal_trainer=False):
        super().__init__(name, age, membership_type)
        self.personal_trainer = personal_trainer

    def request_personal_trainer(self):
        if self.get_membership_status():
            print(f"{self.name}'s personal trainer request accepted.")
            return True
        else:
            print("Cannot request a personal trainer. Membership is not active.")
            return False


def menu():
    print("Welcome to the Fitness Club!")
    
    name = input("Enter a member's name: ")
    age = int(input("Enter member's age: "))
    membership_type = input("Enter membership type (a) Individual b) Family): ").strip().upper()
    
    if membership_type == "FAMILY":
        family_size = int(input("Enter family size: "))
        member = FamilyMember(name, age, membership_type, family_size)
    else:
        member = IndividualMember(name, age, membership_type)
    
    member.register_member()
    
    while True:
        print("\n1. Renew Membership")
        print("2. Cancel Membership")
        print("3. Add Family Member")
        print("4. Request Personal Trainer")
        print("5. Exit")
        
        choice = input("Enter your choice: ").strip()
        
        if choice == '1':
            member.renew_membership()
        elif choice == '2':
            member.cancel_membership()
        elif choice == '3' and isinstance(member, FamilyMember):
            member.add_family_member()
        elif choice == '4' and isinstance(member, IndividualMember):
            member.request_personal_trainer()
        elif choice == '5':
            print("Thank you for visiting the Fitness Club!")
            break
        else:
            print("Invalid choice or operation not available for this membership type.")



menu()


```

```python
'''Event Management
'''

import random

class Event:
    def __init__(self, name, date, time, location):
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []
        self.__eventid = self.__generate_event_id()

    def __generate_event_id(self):
        return random.randint(10000, 99999)

    def get_event_id(self):
        return self.__eventid

    def create_event(self):
        print(f"Event '{self.name}' created successfully!")
        print("Event ID:", self.__eventid)
        print("Date:", self.date)
        print("Time:", self.time)
        print("Location:" ,self.location)

    def add_attendee(self, attendee_name):
        self.__attendees.append(attendee_name)
        print(f"{attendee_name} has been added to the event.")

    def remove_attendee(self, attendee_name):
        if attendee_name in self.__attendees:
            self.__attendees.remove(attendee_name)
            print(f"{attendee_name} has been removed from the event.")
        else:
            print(f"{attendee_name} is not in the attendee list.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def get_attendee_list(self):
        return self.__attendees


class PrivateEvent(Event):
    def __init__(self, name, date, time, location, invitation_required=True):
        super().__init__(name, date, time, location)
        self.invitation_required = invitation_required

    def invite_member(self, attendee_name):
        if self.invitation_required:
            print(f"{attendee_name} has been invited to the private event.")
        else:
            print(f"{attendee_name} can join without an invitation.")


class PublicEvent(Event):
    def __init__(self, name, date, time, location, max_capacity):
        super().__init__(name, date, time, location)
        self.max_capacity = max_capacity

    def add_attendee(self, attendee_name):
        if self.get_total_attendees() < self.max_capacity:
            super().add_attendee(attendee_name)
        else:
            print("Sorry, the event has reached its maximum capacity.")



def menu():
    print("Enter your choice :")
    
    name = input("Enter the event name: ")
    date = input("Enter the event date (YYYY-MM-DD): ")
    time = input("Enter the event time (HH:MM): ")
    location = input("Enter the event location: ")

    event_type = input("Is it a private event (P) or public event (U)? ").upper()

    if event_type == "P":
        event = PrivateEvent(name, date, time, location)
    elif event_type == "U":
        max_capacity = int(input("Enter the maximum capacity for the public event: "))
        event = PublicEvent(name, date, time, location, max_capacity)
    else:
        print("Invalid event type selected.")
        return

    event.create_event()

    while True:
        print("\nOptions:")
        print("1. Add Attendee")
        print("2. Remove Attendee")
        print("3. Show Total Attendees")
        print("4. Show Attendee List")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            attendee_name = input("Enter the attendee's name: ")
            event.add_attendee(attendee_name)
        elif choice == "2":
            attendee_name = input("Enter the attendee's name to remove: ")
            event.remove_attendee(attendee_name)
        elif choice == "3":
            print(f"Total attendees: {event.get_total_attendees()}")
        elif choice == "4":
            print(f"Attendee list: {event.get_attendee_list()}")
        elif choice == "5":
            print("Exiting the system.")
            break
        else:
            print("Invalid choice. Please try again.")


menu()
```

```python
'''Airline Reservation System
'''

import random

class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats):
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = available_seats
        self.__flight_id = self.__generate_flight_id()

    def __generate_flight_id(self):
        return random.randint(10000, 99999)

    def get_flight_id(self):
        return self.__flight_id

    def flight_info(self):
        print(f"Flight Number: {self.flight_number}")
        print(f"Departure: {self.departure_airport} at {self.departure_time}")
        print(f"Arrival: {self.arrival_airport} at {self.arrival_time}")
        print(f"Available Seats: {self.__available_seats}")

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print("Seat booked successfully!")
        else:
            print("No available seats. Booking failed.")

    def cancel_reservation(self):
        self.__available_seats += 1
        print("Reservation canceled successfully!")

    def get_remaining_seats(self):
        return self.__available_seats


class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, meal_included=False):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.meal_included = meal_included

    def meal_service(self):
        if self.meal_included:
            print("Meal service is included in this flight.")
        else:
            print("No meal service available for this flight.")


class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, visa_required=True):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.visa_required = visa_required

    def check_visa_requirement(self):
        if self.visa_required:
            print("Visa is required for this international flight.")
        else:
            print("Visa is not required for this international flight.")



def menu():
    print("Welcome to the Airline Reservation System!")
    
    flight_number = input("Enter the flight number: ")
    departure_airport = input("Enter the departure airport: ")
    arrival_airport = input("Enter the arrival airport: ")
    departure_time = input("Enter the departure time (HH:MM): ")
    arrival_time = input("Enter the arrival time (HH:MM): ")
    available_seats = int(input("Enter the number of available seats: "))

    flight_type = input("Is it a domestic flight (D) or international flight (I)? ").upper()

    if flight_type == "D":
        meal_included = input("Is meal service included? (yes/no): ").lower() == "yes"
        flight = DomesticFlight(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, meal_included)
    elif flight_type == "I":
        visa_required = input("Is a visa required? (yes/no): ").lower() == "yes"
        flight = InternationalFlight(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, visa_required)
    else:
        print("Invalid flight type selected.")
        return

    flight.flight_info()

    while True:
        print("\nOptions:")
        print("1. Book a Seat")
        print("2. Cancel a Reservation")
        print("3. Show Remaining Seats")
        print("4. Flight Specific Service")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            flight.book_seat()
        elif choice == "2":
            flight.cancel_reservation()
        elif choice == "3":
            print(f"Remaining seats: {flight.get_remaining_seats()}")
        elif choice == "4":
            if isinstance(flight, DomesticFlight):
                flight.meal_service()
            elif isinstance(flight, InternationalFlight):
                flight.check_visa_requirement()
        elif choice == "5":
            print("Exiting the system.")
            break
        else:
            print("Invalid choice. Please try again.")


menu()

        
```
