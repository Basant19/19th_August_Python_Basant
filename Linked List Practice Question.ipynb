```python
'''

A doubly linked list is a type of linked list in which each node contains three components:

Data: The value stored in the node.
Next Pointer: A pointer/reference to the next node in the sequence.
Previous Pointer: A pointer/reference to the previous node in the sequence.

'''

class Node:
    def __init__(self, data):
        self.data = data      
        self.next = None      
```

```python
#link list implemntation 

class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
    
    def setdata(self, data):
        self.data = data
    
    def getdata(self):
        return self.data
    
    def setnext(self, next):
        self.next = next
    
    def getnext(self):
        return self.next
    
    def setarbit(self, arbit):
        self.arbit = arbit
    
    def getarbit(self):
        return self.arbit

def traverse(head):
    temp = head 
    while temp:
        print(temp.getdata(), end=" -> ")
        temp = temp.getnext()
    print("None")


head=Node (10)
node2=Node (15)
node3=Node (30)
node4=Node (12)
node5=Node (4)
node6=Node (9)

head.setnext (node2)
node2.setnext (node3)
node3.setnext (node4)
node4.setnext (node5)
node5.setnext (node6)

print("Original list:")
traverse(head)
```

```python
'''reverse a link list in 
'''
  

def reverse(head):
    prev = None
    current = head
    
    while current:
        next_node = current.getnext()  
        current.setnext(prev)           
        prev = current                  
        current = next_node
    
    return prev  


print("Original list:")
traverse(head)

# Reverse the linked list
head = reverse(head)

print("Reversed list:")
traverse(head)

```

```python
'''Merge two sorted linked list into one 
'''

class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
    
    def setdata(self, data):
        self.data = data
    
    def getdata(self):
        return self.data
    
    def setnext(self, next):
        self.next = next
    
    def getnext(self):
        return self.next


def traverse(head):
    temp = head 
    while temp:
        print(temp.getdata(), end=" -> ")
        temp = temp.getnext()
    print("None")



head1 = Node(1)
node2 = Node(1)
node3 = Node(3)
node4 = Node(12)
node5 = Node(40)
node6 = Node(90)

head1.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(node6)


head2=Node(1)
node2 = Node(1)
node3 = Node(3)
node4 = Node(12)
node5 = Node(40)
node6 = Node(90)

head2.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(node6)



def mergelist(head1, head2):
    newnode = Node()
    temp = newnode
    
    while head1 and head2:
        if head1.getdata() < head2.getdata():
            temp.setnext(head1)
            head1 = head1.getnext()
        else:
            temp.setnext(head2)
            head2 = head2.getnext()
        temp = temp.getnext()
    
    if head1:
        temp.setnext(head1)
    elif head2:
        temp.setnext(head2)
    
    return newnode.getnext()

head = mergelist (head1, head2)

print("First list:")
traverse(head1)   

print("Second list:")
traverse(head2) 

print("Merged list:")
traverse(head)
```

```python
'''Write a function to remove nth node from the end in a linked list 
'''


def endremove(head,n):
    fast=head
    for i in range (0,n):
        fast=fast.getnext()
    slow=head
    while fast:
        slow=slow.next
        fast=fast.next
        
    if fast == None:
        return head.next
    
    slow=slow.next.next
    
    return head
        
        
head=Node (10)
node2=Node (15)
node3=Node (30)
node4=Node (12)
node5=Node (4)
node6=Node (9)

head.setnext (node2)
node2.setnext (node3)
node3.setnext (node4)
node4.setnext (node5)
node5.setnext (node6)

print("Original list:")
traverse(head) 

nhead=endremove(head,3)
print("deleted list:")
traverse(nhead) 
```

```python
'''Remove duplicate 
'''

def remove_duplicates(head):
    current = head
    while current and current.getnext():
        if current.getdata() == current.getnext().getdata():
            current.setnext(current.getnext().getnext())
        else:
            current = current.getnext()



head = Node(10)
node2 = Node(15)
node3 = Node(15)
node4 = Node(30)
node5 = Node(30)
node6 = Node(30)

head.setnext(node2)
node2.setnext(node3)
node3.setnext(node4)
node4.setnext(node5)
node5.setnext(node6)

print("Original linked list:")
traverse(head)

remove_duplicates(head)

print("Linked list after removing duplicates:")
traverse(head)
```

```python
'''Find the intersection 
'''

def intersectlinklist (head1,head2):
    temp1=head1
    temp2=head2
    while temp1!=temp2:
        temp1=temp1.getnext() if temp1 else head2
        temp2=temp2.getnext() if temp2 else head1
    return temp1

head1 = Node(1)
node1_2 = Node(1)
node1_3 = Node(3)
node1_4 = Node(12)
node1_5 = Node(40)
node1_6 = Node(90)

head1.setnext(node1_2)
node1_2.setnext(node1_3)
node1_3.setnext(node1_4)
node1_4.setnext(node1_5)
node1_5.setnext(node1_6)


head2 = Node(2)
node2_2 = Node(5)
node2_3 = Node(6)
node2_4 = node1_3
node2_5 = Node(40)
node2_6 = Node(80)

head2.setnext(node2_2)
node2_2.setnext(node2_3)
node2_3.setnext(node2_4)
node2_4.setnext(node2_5)
node2_5.setnext(node2_6)

print ("First Link List")
traverse (head1)

print ("Second Link List")
traverse (head2)

intersect= intersectlinklist(head1,head2)
if intersect:
    print ("Intersection found ")
    traverse(intersect)
    
else:
     print ("Intersection not found")
```

```python
'''Rotate a linked by k positions to the right 
'''

def rotate_right(head, k):
    if not head or k == 0:
        return head
    
    
    length = 1
    last_node = head
    while last_node.next:
        last_node = last_node.next
        length += 1
    
    
    k = k % length
    if k == 0:
        return head
    
   
    last_node.next = head
    
   
    steps_to_new_head = length - k
    new_tail = head
    for _ in range(steps_to_new_head - 1):
        new_tail = new_tail.next
    
    new_head = new_tail.next
    new_tail.next = None
    
    return new_head


def create_linked_list(elements):
    head = Node(elements[0])
    current = head
    for element in elements[1:]:
        current.next = Node(element)
        current = current.next
    return head


def traverse(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")


elements = [1, 2, 3, 4, 5]
k = 2
head = create_linked_list(elements)


print("Original Linked List:")
traverse(head)


rotated_head = rotate_right(head, k)

print(f"Linked List after rotating by {k} places:")
traverse(rotated_head)
```

```python
'''Add two numbers represented by linklist 
'''


def add_two_numbers(l1, l2):
    dummy = Node()  
    current = dummy
    carry = 0

    while l1 or l2 or carry:
       
        val1 = l1.getdata() if l1 else 0
        val2 = l2.getdata() if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10  
        current.setnext(Node(total % 10)) 
        current = current.getnext()

        
        if l1:
            l1 = l1.getnext()
        if l2:
            l2 = l2.getnext()
    
    return dummy.getnext()  
def traverse(head):
    temp = head 
    while temp:
        print(temp.getdata(), end=" -> ")
        temp = temp.getnext()
    print("None")


l1 = Node(2)
l1.setnext(Node(4))
l1.getnext().setnext(Node(3))


l2 = Node(5)
l2.setnext(Node(6))
l2.getnext().setnext(Node(4))

print("First number:")
traverse(l1)

print("Second number:")
traverse(l2)


result = add_two_numbers(l1, l2)

print("Sum of the two numbers:")
traverse(result)
```

```python
'''Clone a linked list with next and Random Pointer 

   Given a linked list of size n 
'''


def clone_linked_list(head):
    if not head:
        return None
    
   
    current = head
    while current:
        new_node = Node(current.getdata())
        new_node.setnext(current.getnext())
        current.setnext(new_node)
        current = new_node.getnext()
    
   
    current = head
    while current:
        current.getnext().setarbit(current.getarbit().getnext() if current.getarbit() else None)
        current = current.getnext().getnext()
    
    
    original = head
    clone = head.getnext()
    clone_head = clone
    
    while original:
        original.setnext(original.getnext().getnext())
        clone.setnext(clone.getnext().getnext() if clone.getnext() else None)
        original = original.getnext()
        clone = clone.getnext()
    
    return clone_head

def traverse_with_arbit(head):
    temp = head 
    while temp:
        arbit_data = temp.getarbit().getdata() if temp.getarbit() else "None"
        print(f"Data: {temp.getdata()}, Arbit: {arbit_data}", end=" -> ")
        temp = temp.getnext()
    print("None")


head = Node(1)
node2 = Node(2)
node3 = Node(3)

head.setnext(node2)
node2.setnext(node3)


head.setarbit(node3)  
node2.setarbit(head)  
node3.setarbit(node2)  

print("Original list with arbit pointers:")
traverse_with_arbit(head)


cloned_head = clone_linked_list(head)

print("Cloned list with arbit pointers:")
traverse_with_arbit(cloned_head)
```
